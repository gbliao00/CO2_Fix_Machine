# Function 

```python
def pump():
    #Commumat with Pump
    ser=serial.Serial('/dev/ttyUSB0', 9600, timeout=0.1)

    #Initial Setting
    ser.write('/1h30001R\r\n'.encode()) 
    time.sleep(CommandWaitTime)
    ser.write('/1h20000R\r\n'.encode()) 
    time.sleep(CommandWaitTime)
    ser.write('/1h10000R\r\n'.encode())
    time.sleep(CommandWaitTime)
#print('Initial Setting OK')

#Velocity Setting
    time.sleep(CommandWaitTime)
#print('Velocilty Setting')
    ser.write('/1S1R\r\n'.encode())
    time.sleep(CommandWaitTime)    
    #print(j)
    #j+=1
    for i in range(2):  # Washing Times
    
        print(i)
        print('Clean Cuvette')
    
    # Turning to Valve
        ser.write('/1h27090R\r\n'.encode())
        time.sleep(CommandWaitTime)
        

    # Output to Pump
        ser.write('/1A100000R\r\n'.encode())
        time.sleep(ActionWaitTime)

    # Turning to Valve4

        ser.write('/1h27150R\r\n'.encode())
        time.sleep(CommandWaitTime)


    #Output to Waste Tank
        time.sleep(CommandWaitTime)
        ser.write('/1A0R\r\n'.encode())
        time.sleep(10)    
        ser.write('/1S1R\r\n'.encode())
        time.sleep(CommandWaitTime)
        ser.write('/1h27040R\r\n'.encode())
        print('Water Input')
        
    # Input
        ser.write('/1A100000R\r\n'.encode())
        time.sleep(ActionWaitTime)
    # Turning to Valve2
        ser.write('/1h27060R\r\n'.encode())
        time.sleep(CommandWaitTime)
    # Output to Cuvette
        time.sleep(CommandWaitTime)
        ser.write('/1A0R\r\n'.encode())
        time.sleep(ActionWaitTime)              
        print('Finish Washing')
        
        # Turning to Valve
        ser.write('/1h27090R\r\n'.encode())
        time.sleep(CommandWaitTime)
        
    # Output to Pump
        ser.write('/1A100000R\r\n'.encode())
        time.sleep(ActionWaitTime)

    # Turning to Valve4
        ser.write('/1h27150R\r\n'.encode())
        time.sleep(CommandWaitTime)

    #Output to Waste Tank
        time.sleep(CommandWaitTime)
        ser.write('/1A0R\r\n'.encode())
        #time.sleep(ActionWaitTime)
        time.sleep(10)

    
    # DAQ Trigger
    GPIO.output(13, GPIO.HIGH)
        
        
    #Lu
    time.sleep(1)
    ser.write('/1h27330R\r\n'.encode())
    time.sleep(1)
    print('Input Lu')
        
    # Input ml
    ser.write('/1A50000R\r\n'.encode())
    time.sleep(ActionWaitTime)
    # Turning to Valve2
    ser.write('/1h27060R\r\n'.encode())
    time.sleep(CommandWaitTime)
    # Output to Cuvette
    time.sleep(CommandWaitTime)
    ser.write('/1A0R\r\n'.encode())
    time.sleep(ActionWaitTime)
    print('Finish')
    ser.close()
    
    
    #Commumat with Pump2
    ser=serial.Serial('/dev/ttyUSB1', 9600, timeout=0.1)
    print('Take ATP')
    
    
    # Turning to Valve
    ser.write('/1h27240R\r\n'.encode())
    time.sleep(CommandWaitTime)
    ser.write('/1A20000R\r\n'.encode())
    time.sleep(5)

    # Initial Setting
    ser.write('/1h30001R\r\n'.encode()) 
    time.sleep(CommandWaitTime)
    ser.write('/1h20000R\r\n'.encode()) 
    time.sleep(CommandWaitTime)
    ser.write('/1h10000R\r\n'.encode())
    time.sleep(CommandWaitTime)
    #print('Initial Setting OK')

    #Turning to Valve
    ser.write('/1h270120R\r\n'.encode())
    time.sleep(1)
    #print('Valve2')    

#  Output to Cuvette
    ser.write('/1A0R\r\n'.encode())
    time.sleep(5)
    ser.close()
    
    #Measurement time
    print('Measurement ATP')
    time.sleep(60)
    
    GPIO.output(13, GPIO.LOW)
    print('Finish')
```

# Function and Run

```python
import RPi.GPIO as GPIO
import time
import serial

GPIO.setmode(GPIO.BOARD)
GPIO.setup(13,GPIO.OUT)  # 13 is GPIO27

#Pump Parameter
CommandWaitTime=1
ActionWaitTime=10

#Waitting Signal for cleaning
j=1
while j<=1:
    print(j)
    # Open Pump OnOff file
    fname = open('PumpOnOff.txt', 'r') 
    for v in iter(fname):
        OnOff=v.strip()             
    fname.close()
    print(OnOff)
    Switch=int(OnOff)
    if Switch==0:
        print('PumpOff')
    else:
        pump()
    j=j+1
```
